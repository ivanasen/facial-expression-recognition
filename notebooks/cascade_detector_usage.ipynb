```python
import cv2

import matplotlib.pyplot as plt
from matplotlib import patches

from models.cascade_face_detector import CascadeFaceDetector
```

```python
%matplotlib inline
```

```python
img = cv2.imread('../data/test_images/group_of_people.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
face_detector = CascadeFaceDetector(scale_factor=1.3, min_neighbours=3)
faces = face_detector.detect(img)
faces
```

```python
figure, ax = plt.subplots(1)
for face in faces:
  rect = patches.Rectangle((face.x, face.y), face.width, face.height, edgecolor='r', facecolor="none")
  ax.add_patch(rect)
ax.imshow(img)
```
